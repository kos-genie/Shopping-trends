```python
#QUESTION 1 How does shopping spending vary based on age and gender?

# Find spending patterns based on age and gender. We have ages from 20 to 50. We can divide into 3 groups: 20-30, 30-40, 40-50.
# Also have 2 groups in gender: Male and Female.
# Box plots will most likely be the best way to compare each group.

# Create a column for age group
conditions = [
    (df_filtered['Age']>= 20) & (df_filtered['Age'] <= 30),
    (df_filtered['Age']>= 31) & (df_filtered['Age'] <= 40),
    (df_filtered['Age']>= 41) & (df_filtered['Age'] <= 50)
]

choices = ['20 to 30', '31 to 40','41 to 50']

df_filtered['Age group'] = np.select(conditions,choices, default = 0)

# df_by_age_gender = df_filtered.groupby(['Age group','Gender']).agg({'Purchase Amount (USD)' : '',
#                                                                     'Category',
#                                                                     'Frequency of Purchases'})

order=['20 to 30', '31 to 40','41 to 50']
sns.set_theme(style="ticks", palette="pastel")
h = sns.boxplot(data=df_filtered,x='Age group',y='Purchase Amount (USD)', hue='Gender',order=order)
h.set_title('Shopping spending by Age group')
h.tick_params(axis='x', rotation=40)
sns.despine(offset=10, trim=True)
sns.move_legend(h, "upper left", bbox_to_anchor=(1, 1))
```

```python
#plot median spending by Age(X-axis) and gender(Hue)
#df_median = df_filtered.groupby(['Age','Gender'])['Purchase Amount (USD)'].median().reset_index()
sns.set_theme(style="ticks", palette="pastel")
# g = sns.lmplot(data=df_filtered, x="Age", y="Purchase Amount (USD)", hue="Gender")
#g.set_title('Median spending by age and gender')

#order=['20 to 30', '31 to 40','41 to 50']
#h = sns.displot(data=df_filtered,x='Purchase Amount (USD)',kind="ecdf",col='Age group', hue='Gender')
h = sns.displot(data=df_filtered, x="Purchase Amount (USD)", col="Gender", kde=True)
#h.set_title('Median spending by age and gender')

#h.tick_params(axis='x', rotation=40)
#sns.despine(offset=10, trim=True)
#sns.move_legend(h, "upper right")
```

```python
#QUESTION 2: Which item types do people pay more for on average?
median_order = df_filtered.groupby('Item Purchased')['Purchase Amount (USD)'].median().sort_values(ascending=False).index

sns.set_theme(style="ticks", palette="pastel")
g = sns.boxplot(data=df_filtered,x='Item Purchased',y='Purchase Amount (USD)',hue='Gender',order=median_order)
g.set_title('Spending by apperal category')
g.tick_params(axis='x', rotation=90)
sns.despine(offset=10, trim=True)
sns.move_legend(g, "upper left", bbox_to_anchor=(1, 1))

```

```python
# QUESTION3: Does shopping spending vary seasonally and geographically? Do people spend more to prepare for winter in colder states/summer in the warm states?

#Compare top 10 states by count
popular_location = df.Location.value_counts().head(10)
print(popular_location)
filtered_df = df[df['Location'].isin(list(popular_location.index))]
h = sns.barplot(data=filtered_df, x="Season", y="Purchase Amount (USD)", estimator="sum",hue='Location', errorbar=None)
h.tick_params(axis='x', rotation=90)
sns.move_legend(h, "upper left", bbox_to_anchor=(1, 1))
h.set_title('Spending (sum) by state/season')
```
